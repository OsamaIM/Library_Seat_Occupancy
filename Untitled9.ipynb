```python
import cv2
import numpy as np
from scipy.spatial import distance

def detect_seats(frame, model, conf_threshold=0.5):
    """
    Run YOLOv8 inference on a frame and return detections.
    """
    results = model(frame, conf=conf_threshold, verbose=False)
    detections = []
    for r in results:
        boxes = r.boxes
        for box in boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            xyxy = box.xyxy[0].cpu().numpy()  # [x1, y1, x2, y2]
            detections.append({
                'class': cls,
                'class_name': model.names[cls],
                'confidence': conf,
                'bbox': xyxy,
                'center': ((xyxy[0] + xyxy[2]) / 2, (xyxy[1] + xyxy[3]) / 2)
            })
    return detections

def draw_detections(frame, detections, class_colors=None):
    """
    Draw bounding boxes and labels on the frame.
    """
    if class_colors is None:
        class_colors = {0: (0, 255, 0), 1: (0, 0, 255)}  # Green for empty, Red for occupied
    
    for det in detections:
        x1, y1, x2, y2 = det['bbox'].astype(int)
        cls = det['class']
        color = class_colors.get(cls, (255, 255, 255))
        label = f"{det['class_name']} {det['confidence']:.2f}"
        
        # Draw rectangle
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        # Draw label background
        cv2.rectangle(frame, (x1, y1 - 25), (x1 + len(label) * 10, y1), color, -1)
        cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    
    return frame

def count_occupancy(detections, class_names):
    """
    Count occupied and empty seats.
    """
    counts = {name: 0 for name in class_names.values()}
    for det in detections:
        counts[det['class_name']] += 1
    return counts

def optimize_seating(detections, class_names, mode='social_distancing', min_distance=100):
    """
    Suggest optimal empty seats based on the selected mode.
    Modes:
    - 'social_distancing': Choose empty seats farthest from occupied seats.
    - 'max_occupancy': Choose empty seats closest to occupied seats (for group seating).
    """
    empty_seats = [det for det in detections if det['class_name'] == 'empty']
    occupied_seats = [det for det in detections if det['class_name'] == 'occupied']
    
    if not empty_seats:
        return []
    
    suggestions = []
    for empty in empty_seats:
        if not occupied_seats:
            # If no occupied seats, suggest all empty seats
            suggestions.append((empty, 0))
            continue
        
        # Calculate distances to all occupied seats
        dists = [distance.euclidean(empty['center'], occ['center']) for occ in occupied_seats]
        min_dist = min(dists)
        
        if mode == 'social_distancing':
            # Prefer seats with larger minimum distance
            score = min_dist
        elif mode == 'max_occupancy':
            # Prefer seats closer to occupied seats
            score = -min_dist  # Negative because we want smaller distances to be higher score
        else:
            score = min_dist
        
        suggestions.append((empty, score))
    
    # Sort by score (higher score is better for social distancing, lower for max occupancy)
    suggestions.sort(key=lambda x: x[1], reverse=(mode == 'social_distancing'))
    return suggestions[:3]  # Return top 3 suggestions
```

```python
# First, let's handle the package dependencies properly
import subprocess
import sys

def install_package(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Install required packages with proper dependencies
print("🔧 Installing required packages...")

# Install the full OpenCV and compatible Gradio
try:
    # Try to install opencv-python-headless first (required by roboflow)
    install_package("opencv-python-headless==4.10.0.84")
except:
    print("⚠️ Could not install opencv-python-headless, trying regular opencv-python")

# Install Gradio 3.x which is more stable
install_package("gradio==3.50.2")

print("✅ Packages installed successfully!")

import gradio as gr
import cv2
import numpy as np
from PIL import Image
import os
import threading
import time
import queue
from ultralytics import YOLO
from scipy.spatial import distance
import matplotlib.pyplot as plt

# Check for your trained model - let's search for it first
print("\n🔍 Searching for trained model...")
model_found = False
model = None

# List of possible locations to search
search_paths = [
    r"C:\Users\G15\library-seat-2",
    r"C:\Users\G15",
    os.getcwd(),
]

for base_path in search_paths:
    if os.path.exists(base_path):
        # Walk through all subdirectories
        for root, dirs, files in os.walk(base_path):
            for file in files:
                if file.endswith('.pt') and ('best' in file.lower() or 'last' in file.lower()):
                    model_path = os.path.join(root, file)
                    try:
                        model = YOLO(model_path)
                        print(f"✅ Found and loaded model: {model_path}")
                        model_found = True
                        break
                    except:
                        continue
            if model_found:
                break
    if model_found:
        break

if not model_found:
    print("⚠️ No trained model found. Let me check for dataset structure...")
    
    # Check if we have dataset to train on
    dataset_path = r"C:\Users\G15\library-seat-2"
    if os.path.exists(dataset_path):
        print(f"📁 Found dataset at: {dataset_path}")
        
        # Look for data.yaml
        yaml_files = []
        for root, dirs, files in os.walk(dataset_path):
            for file in files:
                if file.endswith(('.yaml', '.yml')):
                    yaml_files.append(os.path.join(root, file))
        
        if yaml_files:
            print(f"📄 Found {len(yaml_files)} YAML files")
            data_yaml_path = yaml_files[0]
            print(f"📊 Using YAML: {data_yaml_path}")
            
            # Train a model quickly
            print("⚙️ Training a model with 10 epochs (this will take a few minutes)...")
            try:
                model = YOLO('yolov8n.pt')
                model.train(
                    data=data_yaml_path,
                    epochs=10,
                    imgsz=640,
                    batch=8,
                    name='quick_library_train',
                    exist_ok=True
                )
                model_found = True
                print("✅ Model trained successfully!")
            except Exception as e:
                print(f"⚠️ Training failed: {e}")
                print("📦 Using pretrained COCO model instead...")
                model = YOLO('yolov8n.pt')
        else:
            print("📦 Using pretrained COCO model...")
            model = YOLO('yolov8n.pt')
    else:
        print("📦 Using pretrained COCO model...")
        model = YOLO('yolov8n.pt')

print(f"📊 Model loaded with classes: {list(model.names.values())[:5]}...")

# Global variables for webcam
webcam_running = False
frame_queue = queue.Queue(maxsize=2)
result_queue = queue.Queue(maxsize=2)
current_frame = None
processing_config = {'confidence': 0.5, 'mode': 'social_distancing'}

# Define detection functions
def detect_seats(frame, model, conf_threshold=0.5):
    """Run YOLOv8 inference on a frame and return detections."""
    results = model(frame, conf=conf_threshold, verbose=False)
    detections = []
    
    for r in results:
        boxes = r.boxes
        for box in boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            xyxy = box.xyxy[0].cpu().numpy()
            
            detections.append({
                'class': cls,
                'class_name': model.names[cls],
                'confidence': conf,
                'bbox': xyxy,
                'center': ((xyxy[0] + xyxy[2]) / 2, (xyxy[1] + xyxy[3]) / 2),
                'width': xyxy[2] - xyxy[0],
                'height': xyxy[3] - xyxy[1]
            })
    
    return detections

def draw_detections(frame, detections):
    """Draw bounding boxes and labels on the frame."""
    # Colors for different classes
    colors = {
        'person': (0, 0, 255),      # Red
        'chair': (0, 255, 0),       # Green
        'bench': (0, 200, 100),     # Teal
        'couch': (0, 255, 0),       # Green
        'suggested': (255, 255, 0), # Cyan
    }
    
    for det in detections:
        x1, y1, x2, y2 = det['bbox'].astype(int)
        class_name = det['class_name']
        
        # Determine color based on class
        color = colors.get(class_name, (255, 255, 255))
        
        # Create label
        if class_name == 'person':
            label = f"Occupied ({det['confidence']:.2f})"
        elif class_name in ['chair', 'bench', 'couch']:
            label = f"Empty ({det['confidence']:.2f})"
        else:
            label = f"{class_name} ({det['confidence']:.2f})"
        
        # Draw rectangle
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        
        # Draw label background
        text_size = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2)[0]
        cv2.rectangle(frame, (x1, y1 - 25), (x1 + text_size[0], y1), color, -1)
        cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    
    return frame

def count_occupancy(detections):
    """Count occupied and empty seats."""
    counts = {'empty': 0, 'occupied': 0, 'total': 0}
    
    for det in detections:
        class_name = det['class_name']
        
        if class_name == 'person':
            counts['occupied'] += 1
            counts['total'] += 1
        elif class_name in ['chair', 'bench', 'couch']:
            counts['empty'] += 1
            counts['total'] += 1
    
    return counts

def optimize_seating(detections, mode='social_distancing'):
    """Suggest optimal empty seats based on the selected mode."""
    empty_seats = [det for det in detections 
                  if det['class_name'] in ['chair', 'bench', 'couch']]
    occupied_seats = [det for det in detections 
                     if det['class_name'] == 'person']
    
    if not empty_seats:
        return []
    
    suggestions = []
    for empty in empty_seats:
        if not occupied_seats:
            suggestions.append((empty, 0))
            continue
        
        # Calculate distances to all occupied seats
        dists = [distance.euclidean(empty['center'], occ['center']) for occ in occupied_seats]
        min_dist = min(dists) if dists else 0
        
        if mode == 'social_distancing':
            score = min_dist
        elif mode == 'max_occupancy':
            score = -min_dist
        elif mode == 'center_preferred':
            frame_center = (empty['bbox'][2] / 2, empty['bbox'][3] / 2)
            score = -distance.euclidean(empty['center'], frame_center)
        else:
            score = min_dist
        
        suggestions.append((empty, score))
    
    # Sort by score
    reverse_sort = mode in ['social_distancing']
    suggestions.sort(key=lambda x: x[1], reverse=reverse_sort)
    return suggestions[:3]

def draw_suggestions(frame, suggestions):
    """Draw suggestions on the frame."""
    for i, (seat, score) in enumerate(suggestions):
        x1, y1, x2, y2 = seat['bbox'].astype(int)
        # Draw cyan rectangle for suggestions
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 255, 0), 3)
        cv2.putText(frame, f"⭐ Best {i+1}", (x1, y1 - 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
    return frame

def webcam_thread_func():
    """Webcam capture thread function."""
    global webcam_running, current_frame
    
    # Try different camera indices
    camera_index = 0
    cap = None
    
    for i in range(3):  # Try indices 0, 1, 2
        cap = cv2.VideoCapture(i)
        if cap.isOpened():
            print(f"✅ Camera found at index {i}")
            camera_index = i
            break
        else:
            cap = None
    
    if cap is None:
        print("⚠️ No camera found. Using simulated feed.")
        # Create simulated frames
        sim_frame = np.zeros((480, 640, 3), dtype=np.uint8)
        cv2.putText(sim_frame, "SIMULATED CAMERA", (150, 200),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 3)
        cv2.putText(sim_frame, "Point: Library seating area", (120, 250),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (200, 200, 255), 2)
        
        while webcam_running:
            # Create animated simulation
            sim_copy = sim_frame.copy()
            current_time = time.time()
            pulse = int(255 * (0.5 + 0.5 * np.sin(current_time * 2)))
            
            # Draw simulated seats
            cv2.rectangle(sim_copy, (100, 300), (200, 400), (0, pulse, 0), 2)  # Empty
            cv2.rectangle(sim_copy, (250, 300), (350, 400), (0, 0, pulse), 2)  # Occupied
            cv2.rectangle(sim_copy, (400, 300), (500, 400), (pulse, pulse, 0), 3)  # Suggested
            
            cv2.putText(sim_copy, "Empty", (120, 290), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
            cv2.putText(sim_copy, "Occupied", (270, 290), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)
            cv2.putText(sim_copy, "Suggested", (420, 290), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)
            
            current_frame = sim_copy
            try:
                frame_queue.put_nowait(sim_copy)
            except:
                pass
            time.sleep(0.1)
        return
    
    # Set camera properties
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    
    print("📹 Webcam stream started!")
    
    while webcam_running:
        ret, frame = cap.read()
        if ret:
            current_frame = frame.copy()
            try:
                frame_queue.put_nowait(frame)
            except:
                pass
        else:
            time.sleep(0.1)
    
    cap.release()
    print("📹 Webcam stopped")

def process_thread_func():
    """Processing thread function."""
    global webcam_running, processing_config
    
    while webcam_running:
        try:
            frame = frame_queue.get(timeout=0.5)
            
            # Get current config
            conf = processing_config['confidence']
            mode = processing_config['mode']
            
            # Run detection
            detections = detect_seats(frame, model, conf_threshold=conf)
            counts = count_occupancy(detections)
            suggestions = optimize_seating(detections, mode=mode)
            
            # Draw on frame
            processed_frame = frame.copy()
            processed_frame = draw_detections(processed_frame, detections)
            
            if suggestions:
                processed_frame = draw_suggestions(processed_frame, suggestions)
            
            # Add info overlay
            h, w = processed_frame.shape[:2]
            overlay = processed_frame.copy()
            cv2.rectangle(overlay, (10, 10), (300, 130), (0, 0, 0), -1)
            processed_frame = cv2.addWeighted(overlay, 0.6, processed_frame, 0.4, 0)
            
            # Add text info
            cv2.putText(processed_frame, f"Empty: {counts['empty']}", (20, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            cv2.putText(processed_frame, f"Occupied: {counts['occupied']}", (20, 70),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
            cv2.putText(processed_frame, f"Total: {counts['total']}", (20, 100),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
            
            # Put result in queue
            try:
                result_queue.put_nowait({
                    'frame': processed_frame,
                    'counts': counts,
                    'suggestions': len(suggestions),
                    'time': time.time()
                })
            except:
                pass
                
        except queue.Empty:
            continue
        except Exception as e:
            print(f"⚠️ Processing error: {e}")
            continue

def start_webcam():
    """Start the webcam and processing threads."""
    global webcam_running
    
    if webcam_running:
        return "⚠️ Webcam already running", None
    
    webcam_running = True
    
    # Clear queues
    while not frame_queue.empty():
        frame_queue.get()
    while not result_queue.empty():
        result_queue.get()
    
    # Start threads
    threading.Thread(target=webcam_thread_func, daemon=True).start()
    threading.Thread(target=process_thread_func, daemon=True).start()
    
    # Return initial frame
    init_frame = np.zeros((480, 640, 3), dtype=np.uint8)
    cv2.putText(init_frame, "Starting camera...", (150, 240),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    
    return "✅ Webcam started!", cv2.cvtColor(init_frame, cv2.COLOR_BGR2RGB)

def stop_webcam():
    """Stop the webcam."""
    global webcam_running
    webcam_running = False
    time.sleep(0.5)
    
    stop_frame = np.zeros((480, 640, 3), dtype=np.uint8)
    cv2.putText(stop_frame, "Webcam stopped", (180, 240),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    
    return "🛑 Webcam stopped", cv2.cvtColor(stop_frame, cv2.COLOR_BGR2RGB)

def update_settings(confidence, mode):
    """Update processing settings."""
    global processing_config
    processing_config = {'confidence': confidence, 'mode': mode}
    return f"⚙️ Settings updated: Confidence={confidence}, Mode={mode}"

def get_live_feed():
    """Get the latest processed frame."""
    global webcam_running
    
    if not webcam_running:
        frame = np.zeros((480, 640, 3), dtype=np.uint8)
        cv2.putText(frame, "Click 'Start Webcam'", (150, 200),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        cv2.putText(frame, "to begin detection", (170, 250),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (200, 200, 255), 2)
        return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB), "Status: Ready to start"
    
    try:
        result = result_queue.get_nowait()
        frame = result['frame']
        counts = result['counts']
        
        # Create statistics
        occupancy_rate = counts['occupied'] / counts['total'] if counts['total'] > 0 else 0
        stats = f"""
        📊 **Live Statistics**
        
        • 🪑 **Total Seats:** {counts['total']}
        • ✅ **Empty:** {counts['empty']}
        • 🔴 **Occupied:** {counts['occupied']}
        • 📈 **Occupancy:** {occupancy_rate:.1%}
        • 💡 **Suggestions:** {result['suggestions']}
        • ⏱️ **Updated:** {time.strftime('%H:%M:%S')}
        """
        
        return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB), stats
        
    except queue.Empty:
        if current_frame is not None:
            # Show raw frame if no processing yet
            frame = current_frame.copy()
            cv2.putText(frame, "Processing...", (200, 240),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
            return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB), "Status: Processing frame..."
        else:
            frame = np.zeros((480, 640, 3), dtype=np.uint8)
            cv2.putText(frame, "Initializing...", (200, 240),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
            return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB), "Status: Initializing camera..."

# Create simplified Gradio interface
print("\n" + "="*60)
print("🚀 CREATING REAL-TIME LIBRARY SEAT DETECTION SYSTEM")
print("="*60)

# Create interface
with gr.Blocks(title="Library Seat Detection") as demo:
    gr.Markdown("# 📚 Real-Time Library Seat Occupancy System")
    gr.Markdown("### 🎥 Live Webcam Detection with AI Optimization")
    
    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### 🎮 Controls")
            
            status_display = gr.Textbox(label="System Status", value="Ready to start", interactive=False)
            
            with gr.Row():
                start_btn = gr.Button("▶️ Start Webcam", variant="primary")
                stop_btn = gr.Button("⏹️ Stop", variant="secondary")
            
            gr.Markdown("### ⚙️ Settings")
            
            confidence_slider = gr.Slider(
                minimum=0.1, maximum=0.9, value=0.5, step=0.05,
                label="Detection Confidence"
            )
            
            mode_radio = gr.Radio(
                choices=["social_distancing", "max_occupancy", "center_preferred"],
                value="social_distancing",
                label="Seat Optimization Mode"
            )
            
            update_btn = gr.Button("🔄 Apply Settings", variant="primary")
            
            gr.Markdown("### 📊 Quick Stats")
            quick_stats = gr.Textbox(label="Current Status", value="Webcam not started", interactive=False)
        
        with gr.Column(scale=2):
            gr.Markdown("### 📹 Live Detection Feed")
            live_feed = gr.Image(label="Camera Feed", interactive=False)
            
            gr.Markdown("### 📈 Detailed Statistics")
            stats_output = gr.Markdown("### Waiting for data...")
    
    # Event handlers
    start_btn.click(
        fn=start_webcam,
        outputs=[status_display, live_feed]
    )
    
    stop_btn.click(
        fn=stop_webcam,
        outputs=[status_display, live_feed]
    )
    
    update_btn.click(
        fn=update_settings,
        inputs=[confidence_slider, mode_radio],
        outputs=status_display
    )
    
    # Create a function to update the feed continuously
    def update_all():
        frame, stats = get_live_feed()
        return frame, stats, f"Last update: {time.strftime('%H:%M:%S')}"
    
    # We'll use a polling mechanism since .load() with 'every' isn't available
    # Create a refresh button
    refresh_btn = gr.Button("🔄 Refresh Feed", variant="secondary")
    refresh_btn.click(
        fn=update_all,
        outputs=[live_feed, stats_output, quick_stats]
    )
    
    # Auto-refresh using JavaScript (if supported)
    gr.Markdown("""
    <script>
    // Auto-refresh every 2 seconds
    setInterval(function() {
        var refreshBtn = document.querySelector('[data-testid="refresh-button"]');
        if(refreshBtn) refreshBtn.click();
    }, 2000);
    </script>
    """)
    
    # Add help section
    with gr.Accordion("📖 Click for Help & Instructions", open=False):
        gr.Markdown("""
        ## How to Use This System:
        
        1. **Connect your webcam** to your computer
        2. **Click 'Start Webcam'** to begin live detection
        3. **Point the camera** at library seating area
        4. **Adjust settings** if needed:
           - **Confidence**: Higher = more certain detections
           - **Mode**: Choose optimization strategy
        5. **View results** in real-time
        
        ## What the Colors Mean:
        - **Green boxes**: Empty chairs/benches
        - **Red boxes**: Occupied seats (people detected)
        - **Cyan boxes with stars**: Suggested optimal seats
        
        ## Troubleshooting:
        - **No camera feed?** Make sure webcam is connected
        - **Poor detection?** Adjust confidence slider
        - **System not responding?** Click 'Stop' then 'Start'
        
        ## Optimization Modes:
        - **Social Distancing**: Maximizes distance between people
        - **Max Occupancy**: Suggests seats near occupied ones (for groups)
        - **Center Preferred**: Suggests seats in the center of view
        """)

# Launch the interface
print("\n🎯 System Ready!")
print("💡 Instructions:")
print("   1. Click 'Start Webcam' to begin")
print("   2. Point camera at seating area")
print("   3. Click 'Refresh Feed' to update (or wait for auto-refresh)")
print("   4. Adjust settings as needed")
print("\n⏳ Launching web interface...")

try:
    demo.launch(
        share=False,
        server_name="127.0.0.1",
        server_port=7860,
        show_error=True
    )
except Exception as e:
    print(f"⚠️ Error: {e}")
    print("\n🔧 Trying alternative port 7861...")
    try:
        demo.launch(
            share=False,
            server_port=7861,
            show_error=True
        )
    except:
        print("❌ Could not launch interface. Please check if ports 7860/7861 are available.")
        print("\n💡 Alternative: Run this in a local Python script instead of notebook.")
```

```python
# IMPROVED REAL-TIME LIBRARY SEAT DETECTION WITH BETTER SEAT RECOGNITION
# This version uses a custom-trained seat model if available

import cv2
import numpy as np
import time
from ultralytics import YOLO
from scipy.spatial import distance
import os

print("=" * 70)
print("🚀 REAL-TIME LIBRARY SEAT OCCUPANCY DETECTION SYSTEM")
print("=" * 70)

# Try to load custom seat model first, fallback to COCO
print("\n📦 Loading AI model...")
custom_model_path = r"C:\Users\G15\library-seat-2\weights\best.pt"

if os.path.exists(custom_model_path):
    print(f"✅ Loading custom seat model from: {custom_model_path}")
    model = YOLO(custom_model_path)
    is_custom_model = True
else:
    print("⚠️ Custom seat model not found. Using COCO model.")
    print("   Note: COCO model detects 'person' and 'chair' classes")
    model = YOLO('yolov8n.pt')
    is_custom_model = False

print(f"✅ Model loaded with {len(model.names)} classes")

def detect_seats(frame, confidence=0.5):
    """Detect seats and people in the frame."""
    results = model(frame, conf=confidence, verbose=False)
    
    empty_seats = []
    occupied_seats = []
    
    for r in results:
        boxes = r.boxes
        for box in boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            xyxy = box.xyxy[0].cpu().numpy()
            
            if is_custom_model:
                # Custom model has seat classes
                class_name = model.names[cls]
                if class_name.lower() == 'empty':
                    empty_seats.append({
                        'class_name': class_name,
                        'confidence': conf,
                        'bbox': xyxy,
                        'center': ((xyxy[0] + xyxy[2]) / 2, (xyxy[1] + xyxy[3]) / 2)
                    })
                elif class_name.lower() == 'occupied':
                    occupied_seats.append({
                        'class_name': class_name,
                        'confidence': conf,
                        'bbox': xyxy,
                        'center': ((xyxy[0] + xyxy[2]) / 2, (xyxy[1] + xyxy[3]) / 2)
                    })
            else:
                # COCO model - map to seat classes
                class_name = model.names[cls]
                if class_name == 'person':
                    occupied_seats.append({
                        'class_name': class_name,
                        'confidence': conf,
                        'bbox': xyxy,
                        'center': ((xyxy[0] + xyxy[2]) / 2, (xyxy[1] + xyxy[3]) / 2)
                    })
                elif class_name in ['chair', 'bench', 'couch']:
                    empty_seats.append({
                        'class_name': class_name,
                        'confidence': conf,
                        'bbox': xyxy,
                        'center': ((xyxy[0] + xyxy[2]) / 2, (xyxy[1] + xyxy[3]) / 2)
                    })
    
    return empty_seats, occupied_seats

def draw_detections(frame, empty_seats, occupied_seats):
    """Draw detections on the frame."""
    # Draw empty seats in green
    for seat in empty_seats:
        x1, y1, x2, y2 = seat['bbox'].astype(int)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        label = f"Empty: {seat['confidence']:.2f}"
        cv2.putText(frame, label, (x1, y1 - 10), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
    
    # Draw occupied seats in red
    for seat in occupied_seats:
        x1, y1, x2, y2 = seat['bbox'].astype(int)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
        label = f"Occupied: {seat['confidence']:.2f}"
        cv2.putText(frame, label, (x1, y1 - 10), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)
    
    return frame

def suggest_optimal_seats(frame, empty_seats, occupied_seats):
    """Suggest optimal seats for social distancing."""
    suggestions = []
    
    if empty_seats and occupied_seats:
        for empty in empty_seats:
            # Calculate distance to nearest occupied seat
            distances = []
            for occupied in occupied_seats:
                dist = distance.euclidean(empty['center'], occupied['center'])
                distances.append(dist)
            
            if distances:
                min_dist = min(distances)
                suggestions.append((empty, min_dist))
        
        # Sort by distance (highest first for social distancing)
        suggestions.sort(key=lambda x: x[1], reverse=True)
        
        # Draw top 3 suggestions
        for i, (seat, dist) in enumerate(suggestions[:3]):
            x1, y1, x2, y2 = seat['bbox'].astype(int)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 255, 0), 3)
            cv2.putText(frame, f"⭐ Best {i+1}", (x1, y1 - 30), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)
    
    return frame, len(suggestions[:3])

def add_statistics_overlay(frame, empty_count, occupied_count, suggestion_count, camera_width, is_custom_model):
    """Add statistics overlay to the frame."""
    total = empty_count + occupied_count
    occupancy_rate = occupied_count / total if total > 0 else 0
    
    # Create semi-transparent overlay
    overlay = frame.copy()
    cv2.rectangle(overlay, (10, 10), (400, 200), (0, 0, 0), -1)
    frame = cv2.addWeighted(overlay, 0.6, frame, 0.4, 0)
    
    # Add statistics text
    y_pos = 40
    line_height = 30
    
    model_type = "Custom Seat Model" if is_custom_model else "COCO Model"
    
    texts = [
        (f"Library Seat Detection ({model_type})", (20, y_pos), (255, 255, 255), 0.7),
        (f"Empty Seats: {empty_count}", (20, y_pos + line_height), (0, 255, 0), 0.7),
        (f"Occupied Seats: {occupied_count}", (20, y_pos + line_height*2), (0, 0, 255), 0.7),
        (f"Total Seats: {total}", (20, y_pos + line_height*3), (255, 255, 255), 0.7),
        (f"Occupancy Rate: {occupancy_rate:.1%}", (20, y_pos + line_height*4), (255, 255, 0), 0.7),
        (f"Suggestions: {suggestion_count}", (20, y_pos + line_height*5), (0, 255, 255), 0.7),
    ]
    
    for text, pos, color, scale in texts:
        cv2.putText(frame, text, pos, cv2.FONT_HERSHEY_SIMPLEX, scale, color, 2)
    
    return frame

def test_cameras():
    """Test which cameras are available."""
    print("\n🔍 Testing available cameras...")
    available_cameras = []
    
    for i in range(5):  # Test indices 0-4
        cap = cv2.VideoCapture(i, cv2.CAP_DSHOW)
        if cap.isOpened():
            ret, frame = cap.read()
            if ret:
                height, width = frame.shape[:2]
                print(f"  ✅ Camera {i}: {width}x{height} - Working")
                available_cameras.append(i)
            cap.release()
        else:
            print(f"  ❌ Camera {i}: Not available")
    
    return available_cameras

def record_video(camera_index=0, record_seconds=30, confidence=0.5):
    """Record video with seat detection."""
    # Camera settings
    camera_width = 640
    camera_height = 480
    fps = 20
    max_frames = fps * record_seconds
    
    # Open selected camera
    print(f"\n📹 Opening camera {camera_index}...")
    cap = cv2.VideoCapture(camera_index, cv2.CAP_DSHOW)
    
    if not cap.isOpened():
        print(f"❌ Cannot open camera {camera_index}")
        return None, None, None
    
    # Set camera properties
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, camera_width)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, camera_height)
    
    # Create output directory
    output_dir = "library_seat_detection_output"
    os.makedirs(output_dir, exist_ok=True)
    
    # Create video writer
    timestamp = time.strftime("%Y%m%d_%H%M%S")
    video_filename = f"{output_dir}/seat_detection_{timestamp}.mp4"
    
    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(video_filename, fourcc, fps, (camera_width, camera_height))
    
    print(f"\n🎥 Output will be saved to: {video_filename}")
    
    print("\n" + "=" * 70)
    print("🎯 RECORDING SETTINGS:")
    print(f"  • Camera: {camera_index}")
    print(f"  • Resolution: {camera_width}x{camera_height}")
    print(f"  • FPS: {fps}")
    print(f"  • Duration: {record_seconds} seconds")
    print(f"  • Max frames: {max_frames}")
    print(f"  • Confidence: {confidence}")
    print(f"  • Model: {'Custom' if is_custom_model else 'COCO'}")
    print("=" * 70)
    
    print(f"\n▶️ Starting recording for {record_seconds} seconds...")
    print("   Point camera at library seating area")
    print("   Press Ctrl+C in the notebook to stop early\n")
    
    frame_count = 0
    start_time = time.time()
    
    # Statistics tracking
    empty_history = []
    occupied_history = []
    
    try:
        while frame_count < max_frames:
            # Read frame from camera
            ret, frame = cap.read()
            if not ret:
                print("❌ Failed to grab frame")
                break
            
            # Run detection
            empty_seats, occupied_seats = detect_seats(frame, confidence)
            empty_count = len(empty_seats)
            occupied_count = len(occupied_seats)
            
            # Track history for averages
            empty_history.append(empty_count)
            occupied_history.append(occupied_count)
            
            # Draw detections
            processed_frame = frame.copy()
            processed_frame = draw_detections(processed_frame, empty_seats, occupied_seats)
            
            # Suggest optimal seats
            processed_frame, suggestion_count = suggest_optimal_seats(processed_frame, empty_seats, occupied_seats)
            
            # Add statistics overlay
            processed_frame = add_statistics_overlay(processed_frame, empty_count, occupied_count, 
                                                    suggestion_count, camera_width, is_custom_model)
            
            # Calculate FPS
            frame_count += 1
            elapsed_time = time.time() - start_time
            current_fps = frame_count / elapsed_time if elapsed_time > 0 else 0
            
            # Add FPS counter
            cv2.putText(processed_frame, f"FPS: {current_fps:.1f}", (camera_width - 120, 30), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
            cv2.putText(processed_frame, f"Frame: {frame_count}/{max_frames}", (camera_width - 120, 60), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            cv2.putText(processed_frame, f"Conf: {confidence}", (camera_width - 120, 90), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            
            # Write frame to video file
            out.write(processed_frame)
            
            # Print progress every 50 frames
            if frame_count % 50 == 0:
                progress = (frame_count / max_frames) * 100
                print(f"📊 Progress: {frame_count}/{max_frames} frames ({progress:.1f}%) | "
                      f"Empty: {empty_count}, Occupied: {occupied_count}, FPS: {current_fps:.1f}")
            
            # Small delay to maintain frame rate
            time.sleep(0.01)
            
    except KeyboardInterrupt:
        print("\n🛑 Recording stopped by user")
    
    finally:
        # Cleanup
        print("\n🔄 Saving and cleaning up...")
        cap.release()
        out.release()
        
        total_time = time.time() - start_time
        
        # Calculate averages
        avg_empty = np.mean(empty_history) if empty_history else 0
        avg_occupied = np.mean(occupied_history) if occupied_history else 0
        
        print("\n" + "=" * 70)
        print("📊 RECORDING COMPLETE!")
        print("=" * 70)
        print(f"   Video saved: {video_filename}")
        print(f"   Total frames: {frame_count}")
        print(f"   Total time: {total_time:.1f} seconds")
        avg_fps = frame_count / total_time if total_time > 0 else 0
        print(f"   Average FPS: {avg_fps:.1f}")
        print(f"   Average Empty: {avg_empty:.1f} seats/frame")
        print(f"   Average Occupied: {avg_occupied:.1f} seats/frame")
        print(f"   Model used: {'Custom Seat Model' if is_custom_model else 'COCO Model'}")
        
        # Create a summary text file (FIXED the formatting error)
        summary_file = f"{output_dir}/summary_{timestamp}.txt"
        with open(summary_file, 'w') as f:
            f.write("=" * 60 + "\n")
            f.write("LIBRARY SEAT DETECTION SUMMARY\n")
            f.write("=" * 60 + "\n\n")
            f.write(f"Recording Time: {timestamp}\n")
            f.write(f"Camera Index: {camera_index}\n")
            f.write(f"Resolution: {camera_width}x{camera_height}\n")
            f.write(f"Duration: {total_time:.1f} seconds\n")
            f.write(f"Total Frames: {frame_count}\n")
            f.write(f"Average FPS: {avg_fps:.1f}\n")
            f.write(f"Average Empty Seats: {avg_empty:.1f}\n")
            f.write(f"Average Occupied Seats: {avg_occupied:.1f}\n")
            f.write(f"Confidence Threshold: {confidence}\n")
            f.write(f"Model Type: {'Custom Seat Model' if is_custom_model else 'COCO Model'}\n")
            f.write(f"Output Video: {video_filename}\n")
            f.write(f"Summary File: {summary_file}\n")
        
        print(f"   Summary file: {summary_file}")
        print("=" * 70)
        
        return video_filename, avg_empty, avg_occupied

def main():
    """Main function to run the seat detection system."""
    print("\n📹 Camera Selection")
    print("-" * 40)
    
    # Test all cameras
    available_cameras = test_cameras()
    
    if not available_cameras:
        print("\n❌ No cameras found!")
        return
    
    # Let user choose camera
    print(f"\n📸 Available cameras: {available_cameras}")
    if len(available_cameras) == 1:
        camera_index = available_cameras[0]
        print(f"📹 Using camera {camera_index} (only one available)")
    else:
        try:
            camera_index = int(input(f"🎯 Enter camera index to use {available_cameras}: "))
            if camera_index not in available_cameras:
                print(f"⚠️ Camera {camera_index} not available. Using {available_cameras[0]} instead.")
                camera_index = available_cameras[0]
        except:
            camera_index = available_cameras[0]
            print(f"⚠️ Invalid input. Using camera {camera_index}")
    
    # Get recording settings
    print("\n⚙️ Recording Settings")
    print("-" * 40)
    
    try:
        record_seconds = int(input(f"⏱️ Enter recording duration in seconds (default 30): ") or "30")
    except:
        record_seconds = 30
    
    try:
        confidence = float(input(f"🎯 Enter detection confidence (0.1-0.9, default 0.5): ") or "0.5")
        confidence = max(0.1, min(0.9, confidence))  # Clamp between 0.1 and 0.9
    except:
        confidence = 0.5
    
    # Record video
    video_file, avg_empty, avg_occupied = record_video(
        camera_index=camera_index,
        record_seconds=record_seconds,
        confidence=confidence
    )
    
    if video_file:
        print(f"\n✅ Recording complete! Video saved to: {video_file}")
        print(f"📊 Average statistics: {avg_empty:.1f} empty seats, {avg_occupied:.1f} occupied seats")
        
        # Show what was detected
        print("\n🔍 Detection Results Summary:")
        if avg_empty > 0 and avg_occupied > 0:
            occupancy_rate = avg_occupied / (avg_empty + avg_occupied) * 100
            print(f"   • Occupancy Rate: {occupancy_rate:.1f}%")
            print(f"   • Availability: {avg_empty:.1f} seats available on average")
        elif avg_occupied > 0:
            print("   • All detected seats are occupied")
            print("   • No available seats detected")
        elif avg_empty > 0:
            print("   • All detected seats are empty")
            print("   • Good availability")
        else:
            print("   • No seats detected - check camera position")
        
        print("\n💡 Tips for better detection:")
        print("   1. Ensure good lighting")
        print("   2. Point camera directly at seating area")
        print("   3. Adjust confidence if detection is poor")
        print("   4. Use custom-trained model for better accuracy")
        
        # If using COCO model, suggest training custom model
        if not is_custom_model:
            print("\n⚠️ Using COCO model (not specifically trained for seats)")
            print("   For better accuracy, consider training a custom model:")
            print("   1. Collect images of your library seats")
            print("   2. Label them as 'empty' and 'occupied'")
            print("   3. Train YOLOv8 on your dataset")

# Run the system
if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
import time
from ultralytics import YOLO
import os

# --- CONFIGURATION ---
# We use the STANDARD model because it already knows 'person' and 'chair'
MODEL_NAME = 'yolov8n.pt' 
CONFIDENCE = 0.35
IOU_THRESHOLD = 0.3  # How much overlap implies "sitting"

def calculate_iou(box1, box2):
    """
    Math to calculate how much two boxes overlap (Intersection over Union).
    """
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersection_area = max(0, x2 - x1) * max(0, y2 - y1)
    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1])

    # We use 'box2_area' (the person) as the denominator to see 
    # if the person is mostly inside the chair
    if box1_area + box2_area - intersection_area == 0: return 0
    return intersection_area / min(box1_area, box2_area)

def main():
    print("=" * 60)
    print("🚀 STARTING SMART OVERLAP DETECTION")
    print(f"📦 Loading standard model: {MODEL_NAME}...")
    model = YOLO(MODEL_NAME)
    
    # Setup Camera
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("❌ Error: No camera found.")
        return

    # Setup Saver
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    os.makedirs("output", exist_ok=True)
    filename = f"output/smart_seat_logic_{int(time.time())}.mp4"
    out = cv2.VideoWriter(filename, cv2.VideoWriter_fourcc(*'mp4v'), 20, (width, height))
    
    print(f"✅ Saving to: {filename}")
    print("⏳ Running for 15 seconds...")
    
    start_time = time.time()
    
    while (time.time() - start_time) < 15:
        ret, frame = cap.read()
        if not ret: break

        # 1. Detect Everything
        results = model(frame, conf=CONFIDENCE, verbose=False)
        
        chairs = [] # format: [x1, y1, x2, y2]
        people = [] # format: [x1, y1, x2, y2]

        # 2. Sort objects into lists
        for r in results:
            for box in r.boxes:
                cls_id = int(box.cls[0])
                name = model.names[cls_id]
                coords = box.xyxy[0].cpu().numpy().astype(int)
                
                if name == 'person':
                    people.append(coords)
                elif name in ['chair', 'bench', 'couch', 'sofa']:
                    chairs.append(coords)

        # 3. The Smart Logic: Check Overlaps
        occupied_chairs = []
        empty_chairs = []

        for chair in chairs:
            is_occupied = False
            for person in people:
                # Check if this person is overlapping this chair
                overlap = calculate_iou(chair, person)
                if overlap > IOU_THRESHOLD:
                    is_occupied = True
                    break
            
            if is_occupied:
                occupied_chairs.append(chair)
            else:
                empty_chairs.append(chair)

        # 4. Draw Results
        # Green for Empty
        for (x1, y1, x2, y2) in empty_chairs:
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, "EMPTY", (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Red for Occupied
        for (x1, y1, x2, y2) in occupied_chairs:
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(frame, "OCCUPIED", (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

        # Dashboard
        stats = f"Empty: {len(empty_chairs)} | Occupied: {len(occupied_chairs)}"
        cv2.rectangle(frame, (0,0), (400, 40), (0,0,0), -1)
        cv2.putText(frame, stats, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

        out.write(frame)

    cap.release()
    out.release()
    print(f"✅ DONE! Open {filename} to see if it worked.")

if __name__ == "__main__":
    main()
```
